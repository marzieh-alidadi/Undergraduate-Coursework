```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
df_train = pd.read_csv('Prepared_Train.csv')
df_train
```

```python
df_test = pd.read_csv('Prepared_Test.csv')
df_test
```

```python
df_train.dtypes
```

the date column is object and we need to change it to datetime:

```python
df_train['date'] = pd.to_datetime(df_train.date , format = '%Y-%m-%d')
data_train = df_train.drop(['date'], axis=1)
data_train.index = df_train.date
data_train
```

```python
df_test['date'] = pd.to_datetime(df_test.date , format = '%Y-%m-%d')
data_test = df_test.drop(['date'], axis=1)
data_test.index = df_test.date
data_test
```

fit the model

```python
model = VAR(endog = data_train)
model_fit = model.fit()
```

make prediction on test dataset:

```python
prediction = model_fit.forecast(model_fit.y, steps=len(data_test))
```

The predictions are in the form of an array, where each list represents the predictions of the row. 

We will transform this into a more presentable format:

```python
prediction
```

converting predictions to dataframe:

```python
pred = pd.DataFrame(index=range(0,len(prediction)),columns=[data_test.columns])
for j in range(0,20):
    for i in range(0, len(prediction)):
        pred.iloc[i][j] = prediction[i][j]
```

```python
pred
```

check rmse & mae & mse:

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

for i in data_test.columns:
    print('rmse value for', i, 'is : ', np.sqrt(mean_squared_error(pred.loc[:, [i]], data_test.loc[:, [i]])))
    print('mae value for', i, 'is : ', mean_absolute_error(pred.loc[:, [i]], data_test.loc[:, [i]]))
    print('mse value for', i, 'is : ', mean_squared_error(pred.loc[:, [i]], data_test.loc[:, [i]]))
    print('\n')
```
