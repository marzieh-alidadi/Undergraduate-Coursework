# Merge Train and Test 

```python

import pandas as pd
df = pd.read_csv('train.csv')

df_test = pd.read_csv('test_timeseries.csv')
df_test
```

```python
df_test = df_test.dropna()
```

```python
df_test.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler()
scaler2 = StandardScaler()

df_to_normalize_train = df[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]

df_to_normalize_test = df_test[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]

scaler1.fit(df_to_normalize_train)
df_to_normalize_train = scaler1.transform(df_to_normalize_train)

df_normal_train = pd.DataFrame(df_to_normalize_train)
df_normal_train.columns = ['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']


scaler2.fit(df_to_normalize_test)
df_to_normalize_test = scaler2.transform(df_to_normalize_test)

df_normal_test = pd.DataFrame(df_to_normalize_test)
df_normal_test.columns = ['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']
```

```python
df_total = {}
df_total['date'] = pd.to_datetime(df['date'])
df_total['score'] = df['score'].astype(int)

df_total = pd.DataFrame(df_total)

df_total = pd.concat([df_total,df_normal_train],axis=1)
df_total.set_index('date',inplace=True)
df_total
```

```python
df_total2 = {}
df_total2['date'] = pd.to_datetime(df_test['date'])
df_total2['score'] = df_test['score'].astype(int)

df_total2 = pd.DataFrame(df_total2)
df_total2.reset_index(drop=True, inplace=True)
df_total2 = pd.concat([df_total2,df_normal_test],axis=1)
df_total2.set_index('date',inplace=True)
df_total2
```

```python
df = pd.concat([df_total,df_total2],axis=0)
df
```

```python
df.to_csv('merge_dataset.csv')
```

# PCA

```python
from sklearn.decomposition import PCA

pca01 = PCA(n_components=5)
principComp = pca01.fit_transform(df_total)
```

```python
import numpy as np
np.cumsum(pca01.explained_variance_ratio_)
```

# DecisionTreeClassifier

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

principComp = pd.DataFrame(principComp)

X_train, X_test, y_train, y_test = train_test_split(principComp, df_total['score'], test_size=0.25, random_state=1)


clf4 = DecisionTreeClassifier()
clf4 = clf4.fit(X_train,y_train)
```

```python
y_pred = clf4.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_pred, digits=3))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_pred, digits=3))
```

# KNeighborsClassifier Model 

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(metrics.classification_report(y_test, y_pred, digits=3))
```
