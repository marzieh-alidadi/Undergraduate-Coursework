```python
import pandas as pd

df = pd.read_csv('timeSeries_v2.csv')
# convert data frame to timeSeries
df['date'] = pd.to_datetime(df['date'])
df.set_index('date',inplace=True)
df
```

```python
df.corr().score.sort_values()
```

```python
df_new = df[['PS','T2M_RANGE','TS','T2M_MAX','T2M','score']]
df_new
```

```python
df_shift= df_new[['TS','T2M_MAX','T2M_RANGE','T2M','PS']].shift()
df_shift.columns = ['TS_pre','T2M_MAX_pre','T2M_RANGE_pre','T2M_pre','PS_pre']
df_total = pd.concat([df_new,df_shift],axis=1)
df_total
```

```python
df_total.fillna(value=0,inplace=True)
df_total
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_total[['TS','T2M_MAX','TS_pre','T2M_MAX_pre','T2M_RANGE','T2M_RANGE','T2M','T2M_pre','PS','PS_pre']], df_total['score'], test_size=0.2, random_state=1)

clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```

```python
y_pred = clf2.predict(X_test)
```

```python
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
import pandas as pd
df = pd.read_csv('train.csv')

df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_to_normalize = df[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
scaler.fit(df_to_normalize)
df_to_normalize = scaler.transform(df_to_normalize)

df_normal = pd.DataFrame(df_to_normalize)
df_normal.columns = ['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']
df_normal
```

```python
df_total = {}
df_total['date'] = pd.to_datetime(df['date'])
df_total['score'] = df['score'].astype(int)

df_total = pd.DataFrame(df_total)

df_total = pd.concat([df_total,df_normal],axis=1)
df_total.set_index('date',inplace=True)
df_total
```

```python
df_shift = {}
for i in df_total.columns:
    df_shift[i + '_pre']= df_total[i].shift()
df_shift = pd.DataFrame(df_shift)
df_total = pd.concat([df_total,df_shift],axis=1)
df_total
```

```python
df_total.fillna(value=0,inplace=True)
```

```python
from sklearn.decomposition import PCA

pca01 = PCA(n_components=5)
principComp = pca01.fit_transform(df_total)
```

```python
import numpy as np
np.cumsum(pca01.explained_variance_ratio_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

principComp = pd.DataFrame(principComp)

X_train, X_test, y_train, y_test = train_test_split(principComp, df_total['score'], test_size=0.25, random_state=1)


clf4 = DecisionTreeClassifier()
clf4 = clf4.fit(X_train,y_train)
```

```python
y_pred = clf4.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```
