```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
df3 = pd.read_csv('train.csv')

df3
```

```python
df3['score'] = df3['score'].astype(int)
```

```python
class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count = df3.score.value_counts()
print(class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count)
```

```python
class_0 = df3[df3['score'] == 0]
class_1 = df3[df3['score'] == 1]
class_2 = df3[df3['score'] == 2]
class_3 = df3[df3['score'] == 3]
class_4 = df3[df3['score'] == 4]
class_5 = df3[df3['score'] == 5]
```

```python
class_0_under = class_0.sample(class_1_count)
class_2_under = class_2.sample(class_1_count,replace=True)
class_3_under = class_3.sample(class_1_count,replace=True)
class_4_under = class_4.sample(class_1_count,replace=True)
class_5_under = class_5.sample(class_1_count,replace=True)

df = pd.concat([class_0_under,class_1,class_2_under,class_3_under,class_4_under,class_5_under], axis=0)
df
```

```python
class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count = df.score.value_counts()
print(class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_to_normalize = df[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
scaler.fit(df_to_normalize)
df_to_normalize = scaler.transform(df_to_normalize)
df_to_normalize
```

```python
df_normal = pd.DataFrame(df_to_normalize)
df_normal.columns = ['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']
df_normal
```

```python
df_score = df[['score']]
df_score = df_score.reset_index(drop=True)
df_score
```

```python
df_normal = df_normal.loc[df_normal.index.drop_duplicates(keep=False),:]
df_score = df_score.loc[df_score.index.drop_duplicates(keep=False),:]
df_score
```

```python
df_ready = pd.concat([df_normal, df_score], axis=1, join='outer')
df_ready
```

```python
a = df_ready.corr()
a.sort_values(by='score',ascending=False)
```

```python
x_train = df_ready[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
y_train = df_ready[['score']]
```

```python
import pandas as pd
df_test = pd.read_csv('test_timeseries.csv')
df_test = df_test.dropna()

df_test['score'] = df_test['score'].astype(int)
```

```python
y_test = df_test[['score']]
df_test = df_test[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                              ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]

scaler = StandardScaler()
scaler.fit(df_test)
df_test = scaler.transform(df_test)


df_test = pd.DataFrame(df_test)
df_test.columns = ['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']
df_test
```

```python
x_test = df_test[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
```

```python
from sklearn.neural_network import MLPClassifier
import numpy as np

y_train_2 = y_train.values.ravel()

clf = MLPClassifier(random_state=1, max_iter=300).fit(x_train, y_train_2)
y_pred = clf.predict(x_test)
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred.columns = ['score']


from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred, labels=[1,2], average='micro')
precision = precision_score(y_test, y_pred, labels=[1,2], average='micro')
recall
```
