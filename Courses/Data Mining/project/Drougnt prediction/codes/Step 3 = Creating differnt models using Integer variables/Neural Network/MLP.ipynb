```python
x_train = df5[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
y_train = df5[['score']]
```

```python
df_test = pd.read_csv('test_timeseries.csv')
df_test = df_test.dropna()

scaler = StandardScaler()
y_test = df_test[['score']]
df_test = df_test[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
scaler.fit(df_test)

df_test = scaler.transform(df_test)


df_test = pd.DataFrame(df_test)
df_test.columns = ['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']
df_test
```

```python
x_test = df_test[['PRECTOT','PS','QV2M','T2M','T2MDEW','T2MWET','T2M_MAX',
                               'T2M_MIN','T2M_RANGE','TS','WS10M','WS10M_MAX','WS10M_MIN','WS10M_RANGE'
                                ,'WS50M','WS50M_MAX','WS50M_MIN','WS50M_RANGE']]
```

```python
from sklearn.neural_network import MLPClassifier
import numpy as np

y_train_2 = y_train.values.ravel()
y_train_2 =y_train_2.astype('int')
y_train_2
```

```python
clf = MLPClassifier(random_state=1, max_iter=300).fit(x_train, y_train_2)
y_pred = clf.predict(x_test)
```

```python
y_pred = pd.DataFrame(y_pred)
y_pred.columns = ['score']
```

```python
from sklearn.metrics import accuracy_score
y_test =y_test.astype('int')
print(accuracy_score(y_test, y_pred))
```

```python
y_pred[y_pred['score'] != 0]
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred, labels=[1,2], average='micro')
precision = precision_score(y_test, y_pred, labels=[1,2], average='micro')
recall
```
