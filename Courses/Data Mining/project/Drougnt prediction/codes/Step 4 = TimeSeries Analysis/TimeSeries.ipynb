# ARIMA model

```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
```

train and test sets:

```python
train_1001 = df_1001.values
test_1001 = df_1001_test.values

history_1001 = [x for x in train_1001]
predictions_1001 = list()
```

```python
# walk-forward validation
for t in range(len(test_1001)):
    model = ARIMA(history_1001, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions_1001.append(yhat)
    obs = test_1001[t]
    history_1001.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
    
# evaluate forecasts
rmse = sqrt(mean_squared_error(test_1001, predictions_1001))
print('Test RMSE: %.3f' % rmse)

# plot forecasts against actual outcomes
pyplot.plot(test_1001,color='green')
pyplot.plot(predictions_1001, color='red')
pyplot.show()
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
```

```python
predictions_1001 = pd.DataFrame(predictions_1001)

predictions_1001 = predictions_1001.astype(int)
predictions_1001
```

```python
accuracy = accuracy_score(test_1001, predictions_1001)
recall = recall_score(test_1001, predictions_1001, labels=[1,2], average='micro')
precision = precision_score(test_1001, predictions_1001, labels=[1,2], average='micro')
```
