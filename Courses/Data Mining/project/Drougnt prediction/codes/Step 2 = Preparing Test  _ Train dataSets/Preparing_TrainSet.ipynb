```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('Train.csv')

df['date'] = pd.to_datetime(df['date'])
df.set_index('date',inplace=True)
df
```

```python
df['score'] = df['score'].astype(int)
df
```

```python
class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count = df.score.value_counts()
print(class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count)
```

```python
class_0 = df[df['score'] == 0]
class_1 = df[df['score'] == 1]
class_2 = df[df['score'] == 2]
class_3 = df[df['score'] == 3]
class_4 = df[df['score'] == 4]
class_5 = df[df['score'] == 5]
```

```python
class_0_under = class_0.sample(class_1_count)
class_2_under = class_2.sample(class_1_count,replace=True)
class_3_under = class_3.sample(class_1_count,replace=True)
class_4_under = class_4.sample(class_1_count,replace=True)
class_5_under = class_5.sample(class_1_count,replace=True)

df = pd.concat([class_0_under,class_1,class_2_under,class_3_under,class_4_under,class_5_under], axis=0)
df.sort_values('date')
```

```python
class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count = df.score.value_counts()
print(class_0_count,class_1_count,class_2_count,class_3_count,class_4_count,class_5_count)
```

```python
df = df.sort_values('date')
df
```
