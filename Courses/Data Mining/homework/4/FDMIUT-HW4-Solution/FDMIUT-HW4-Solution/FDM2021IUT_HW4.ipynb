```python
# Q1

import pandas as pd
from sklearn.model_selection import train_test_split
from  sklearn import preprocessing

Vehicle = pd.read_csv('Vehicle.csv', sep=',')
Vehicle_df = pd.DataFrame(Vehicle)

Vehicle_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

Vehicle_df.iloc[:,-1] = le.fit_transform(Vehicle_df.iloc[:,-1])
Vehicle_df.head()
```

```python
col = Vehicle_df.columns
min_max_scaler = preprocessing.MinMaxScaler()
Vehicle_df.iloc[:,:-1] = min_max_scaler.fit_transform(Vehicle_df.iloc[:,:-1])
Vehicle_df = pd.DataFrame(Vehicle_df)
Vehicle_df.columns=col
Vehicle_df
```

```python
import seaborn as sns

sns.heatmap(Vehicle_df.corr(), cmap="BrBG")
```

```python
X = Vehicle_df.iloc[:,:-1]
X
```

```python
Y = Vehicle_df.iloc[:,-1]
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size = 0.2, random_state=50)
```

```python
y_train.hist()
```

```python
y_test.hist()
```

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTree = DecisionTreeClassifier(criterion='entropy')
DecisionTree.fit(x_train, y_train)
```

```python
y_pred = DecisionTree.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy = ", accuracy)
```

```python

Diabetes_data = pd.read_csv('Diabetes.csv', sep=',', 
                           names = ['pregnant','Plasma glucose'
                               ,'Diastolic blood pressure','Triceps skinfold thickness'
                               ,'serum insulin','Body mass index'
                               ,'Diabetes pedigree','Age','Class'])
Diabetes_data_df = pd.DataFrame(Diabetes_data)

head = Diabetes_data_df.head()
head
```

```python
col = Diabetes_data_df.columns
min_max_scaler = preprocessing.MinMaxScaler()
Diabetes_data_df = min_max_scaler.fit_transform(Diabetes_data_df)
Diabetes_data_df = pd.DataFrame(Diabetes_data_df)
Diabetes_data_df.columns=col
Diabetes_data_df
```

```python
X = Diabetes_data_df.iloc[:,:-1]
X
```

```python
Y = Diabetes_data_df.iloc[:,-1]
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size = 0.2, random_state=50)
```

```python
y_train.hist()
```

```python
y_test.hist()
```

```python
from sklearn.ensemble import RandomForestClassifier

RandomForest = RandomForestClassifier(max_depth = 3, criterion='entropy')
RandomForest.fit(x_train, y_train)
```

```python
y_pred = RandomForest.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)*100
print("Accuracy = ", accuracy)
```

```python
accuracy_list = []
for i in range(1,15):
    RandomForest = RandomForestClassifier(max_depth = i, criterion='entropy')
    RandomForest.fit(x_train, y_train)
    y_pred = RandomForest.predict(x_test)
    accuracy_list.append(accuracy_score(y_test,y_pred)*100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
plt.xticks(range(0,15),range(1,15))
plt.plot(accuracy_list, color="blue", linewidth=2)
plt.xlabel("depth",c="green")
plt.ylabel("accuracy",c="green")
plt.yticks(c= 'green')
plt.xticks(c= 'green')
plt.title("Random Forest",c="green")
plt.show()
```
