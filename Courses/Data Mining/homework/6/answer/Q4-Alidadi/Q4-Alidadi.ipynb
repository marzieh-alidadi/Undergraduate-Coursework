```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend import preprocessing
```

```python
## part b:
```

```python
bs=pd.read_csv('Desktop/Basket.csv',names=['ITEMS'],delimiter='/')
bs
```

```python
df=pd.DataFrame(bs)
df=pd.DataFrame(df,columns=['INDEX','ITEMS'])
for i in range(0,12526):
    df['INDEX'][i]=i+1
df['INDEX']=df['INDEX'].astype('int64')
```

```python
df['ITEMS'] = df['ITEMS'].str.strip(',')
df
```

```python
df['ITEMS']=df['ITEMS'].str.replace(",", " ")
df['ITEMS']=df['ITEMS'].str.split()
df
```

```python
item_list = []
for i in df['INDEX']:
    item_list.append(df['ITEMS'][i-1])
print(item_list[0:5])
```

```python
online_encoder = preprocessing.TransactionEncoder()
online_encoder_array = online_encoder.fit_transform(item_list)
```

```python
online_encoder_df = pd.DataFrame(online_encoder_array, columns=online_encoder.columns_)
online_encoder_df
```

```python
## part c:
```

```python
frequent_itemsets = apriori(online_encoder_df, min_support=0.1, use_colnames=False)
frequent_itemsets
```
