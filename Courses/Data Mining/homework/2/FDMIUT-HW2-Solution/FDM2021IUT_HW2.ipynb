```python
# Q2
def isfloat(x):
    try:
        a = float(x)
    except (TypeError, ValueError):
        return False
    else:
        return True

def isint(x):
    try:
        a = float(x)
        b = int(a)
    except (TypeError, ValueError):
        return False
    else:
        return a == b
    
def printNA(x):
    if not(isint(x) or isfloat(x)):
        print(x)
```

```python
# 5
# a
df = pd.read_csv('Diabetes.csv', sep=',',na_values={"?","MISS","''"})

from sklearn.impute import SimpleImputer
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")
plt.figure(figsize=(20,20))

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

imputer = imputer.fit(df.iloc[:,:8])
df.iloc[:,:8] = imputer.transform(df.iloc[:,:8])
```

```python
# b
sns.boxplot(data=df, orient="w",width=0.6, palette="colorblind")
```

```python
# b

def Outlier(col_data):
    Q1 = np.percentile(col_data, 25)
    Q3 = np.percentile(col_data, 75)
    IQR = Q3 - Q1
    Low_bound = Q1 - (1.5 * IQR)
    High_bound = Q3 + (1.5 * IQR)

    return Low_bound, High_bound

for i in df.columns:
    print(i +  ' Outliers:\n')
    col = df[i]

    Low_bound, High_bound = Outlier(col)
    for j in col:
        if (j > High_bound):
            print('high outlier ' + str(j))
        elif (j < Low_bound):
            print('low outlier ' + str(j))
    print("\n")
```
