```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LogisticRegression, PoissonRegressor
from sklearn.datasets import load_breast_cancer, load_boston
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MultiLabelBinarizer

from mlxtend.frequent_patterns import apriori, association_rules

import jenkspy
from gsppy.gsp import GSP
```

```python
# Q1
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df['cancer'] = data.target
df
```

```python
df = df.dropna()
for i in df.columns:
    if i != 'cancer':
        df[i] = (df[i] - df[i].mean())/df[i].std()
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1:], test_size = 0.2, random_state=42)
```

```python
lg_model = LogisticRegression(random_state=25).fit(x_train, y_train)
```

```python
lg_model.score(x_test, y_test)
```

```python
print('liblinear')
lglin_model = LogisticRegression(random_state=25, solver='liblinear').fit(x_train, y_train)
print('train: ',lglin_model.score(x_train, y_train),', test: ',lglin_model.score(x_test, y_test))

print('saga')
lgsaga_model = LogisticRegression(random_state=25, solver='saga').fit(x_train, y_train)
print('train: ',lgsaga_model.score(x_train, y_train),', test: ',lgsaga_model.score(x_test, y_test))
```

```python
boston_data = load_boston()
boston_data_df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
boston_data_df['Price'] = boston_data.target
boston_data_df.head()
```

```python
col = boston_data_df.columns
min_max_scaler = preprocessing.MinMaxScaler()
boston_data_df = min_max_scaler.fit_transform(boston_data_df)
boston_data_df = pd.DataFrame(boston_data_df)
boston_data_df.columns=col
boston_data_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(boston_data_df.iloc[:,:-1], boston_data_df.iloc[:,-1:],test_size = 0.2, random_state=42)
```

```python
po_model = PoissonRegressor().fit(x_train, y_train)
y_pred = po_model.predict(x_test)
```

```python
print('mean_absolute_error: ', mean_absolute_error(y_pred, y_test))
print('mean_squared_error: ', mean_squared_error(y_pred, y_test))
print('root_mean_squared_error: ', np.sqrt(mean_squared_error(y_pred, y_test)))
```

```python
df = pd.read_csv('Sequence.csv',header = None,names=range(7))
df = df.fillna(0)
transactions = []
ds = df.values
for i in range(len(df)):
    temp = []
    for j in range(7):
        if ds[i, j] != 0:
            temp.append(ds[i, j])
    transactions.append(temp)
df
```

```python
result = GSP(transactions).search(0.3)
result
```

```python
specific_buy = []

for x in transactions:
    before_sell = ''
    for y in x:
        if(before_sell == 'Bread' and y == 'Sweet'):
            specific_buy.append(x)
        before_sell = y
```

```python
df = pd.read_csv('Heart.csv')
df = df.dropna()
df
```

```python
breaks = jenkspy.jenks_breaks(df['maximum heart rate achieved'],nb_class = 5)
df1 = pd.DataFrame(df[['resting electrocardiographic', 'chest pain type', 'exercise induced angina', 'target']])
df1['max hr bin'] = pd.cut(x = df['maximum heart rate achieved'], 
                           bins = breaks, right = False, labels = ['hr in (71,116)','hr in (116,136)',
                                                                   'hr in (136,153)','hr in (153,169)',
                                                                   'hr in (169,202)'])
df1 = pd.get_dummies(df1, columns = ['resting electrocardiographic', 'chest pain type',
                                          'exercise induced angina', 'max hr bin', 'target'])
df1
```

```python
frequent_itemsets = apriori(df1, min_support=0.25, use_colnames=True)
pd.options.display.max_colwidth = 500
frequent_itemsets.tail(5)
```

```python
rules = association_rules(frequent_itemsets, metric='lift')
rules = pd.concat([rules[rules.consequents == frozenset({'target_1'})],
                   rules[rules.consequents == frozenset({'target_0'})]], ignore_index=True)
rules
```
